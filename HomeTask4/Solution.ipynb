#### Детектор зарегестрировал N событий, фон составил M событий. Какая статистическая погрешность?

Счетчик событий обычно представляется с помощью распределения Пуассона.

Полагаю что события происходят с некоторой интенсивностью.

Также из условия дано что события независимы друг от друга.

Число частиц которое мы хотим померить:
$K = N - M$

$D[K]$ = $D[N-M]$ = $E((N-M)^2) - (E(N-M))^2 = E(N^2 - 2NM + M^2) - (E(N)^2 - 2E(N)E(M) + E(M)^2)$

$= E(N^2) - E(N)^2 + E(M^2) - E(M)^2 $.

А это значит что $\sigma[K] = \sqrt{N+M}$

#### Пусть вы провели измерение σ, потом вы провели второе такое измерение. Как ошибка среднего зависит от корреляции между измерениями?

Самая важная тут формула:

$D[X+Y] = D[X] + D[Y] + 2*COV[X,Y]$
$COV[X,Y] = \rho \sigma^2$

$D[X+Y] = 2 \sigma^2 + 2 \rho \sigma^2 = 2\sigma^2(1 + \rho)$
$D[\frac{X+Y}{2}] = \frac{1}{2^2} D[X+Y] = \frac{\sigma^2(1+\rho)}{2}$
