#### Дайте определение систематической ошибки

Систематическая ошибка - тип ошибки связанный с выбором метода измерения при которой результат отличается от истинной величины

#### Как будут изменяться систематическая и статистическая ошибки с ростом статистики?

Систематическая - не изменяется
Статистическая - уменьшается по закону больших чисел ($\frac{1}{\sqrt{N}}$)

#### Как можно улучшить систематические погрешности?

Не всегда систематическая погрешность - это ужасно.
К примеру если это отклонение постоянное, мы можем просто сдвинуть измерения $X^{true} = X^{measured} + \textbf{bias}_{const}$.

Также можно понять причины по которым возникает систематическое смещение, к примеру это может быть неправильно поставленный эксперимент, смещенная выборка.

Примером того чего можно добиться узнав причины, является введение балансирующих коэффициентов.

#### В соответствии с рекомендациями лабника вы привели исходную зависимость к линейной, как при этом изменятся ошибки?

Изменится систематическая ошибка к примеру если реальная зависимость - квадратичная.

Подобный вид ошибки можно легко заметить, если построить график и убедиться что ошибка имеет квадратичную зависимость

#### Детектор зарегестрировал N событий, фон составил M событий. Какая статистическая погрешность?

Счетчик событий обычно представляется с помощью распределения Пуассона.

Полагаю что события происходят с некоторой интенсивностью.

Также из условия дано что события независимы друг от друга.

Число частиц которое мы хотим померить:
$K = N - M$

$D[K]$ = $D[N-M]$ = $E((N-M)^2) - (E(N-M))^2 = E(N^2 - 2NM + M^2) - (E(N)^2 - 2E(N)E(M) + E(M)^2)$

$= E(N^2) - E(N)^2 + E(M^2) - E(M)^2 $.

А это значит что $\sigma[K] = \sqrt{N+M}$

#### Пусть вы провели измерение σ, потом вы провели второе такое измерение. Как ошибка среднего зависит от корреляции между измерениями?

Самая важная тут формула:

$D[X+Y] = D[X] + D[Y] + 2*COV[X,Y]$
$COV[X,Y] = \rho \sigma^2$

$D[X+Y] = 2 \sigma^2 + 2 \rho \sigma^2 = 2\sigma^2(1 + \rho)$
$D[\frac{X+Y}{2}] = \frac{1}{2^2} D[X+Y] = \frac{\sigma^2(1+\rho)}{2}$

### Исследуется асимметрия вперед-назад, эксперимент заканчивается после набора ...

$N = F + B$, значит мы не можем утверждать что $F, B$ - независимые величины.

Посмотрим на вероятность получения конфигурации системы
$P(F=f; B=b)$. Формально это можно записать как:

$P(F = f \cap B = b) = \; ?$.

Так как данные события происходят при наборе величины $N$, ее следует добавить (с учетом того что сумма конфигурации дает $N$)

$P(F = f \cap B = b | f + b = N) = \; ?$.

Тут нужна формула Байеса:

$P(A|B) = \frac{P(B|A)P(A)}{P(B)}$

В нашем случае это превратится в:

$P(F = f \cap B = b | f + b = N) = \frac{P(f+b=N | F= f \cap B = b)P( F= f \cap B = b)}{P(f + b = N)}$

$g$ - распределение суммы случайных величин $F + B$

Распределение $F = \frac{p^f_F p^{N-f}_B}{g^N[F,B]} = \frac{p^f_F p^{N-f}_B}{g[p^f_F p^b_B]}$

$r = \frac{F-B}{F + B} = \frac{F - N + F}{F + N - F} = \frac{2F - N}{N} = \frac{2F}{N} - 1$

$\frac{\partial r}{\partial F} = \frac{2}{N}$

$\sigma_{assym} = (\frac{\partial r}{\partial F})\sigma_F$

$\sigma_{assym}^2 = \frac{4}{N^2}\sigma_F^2$
